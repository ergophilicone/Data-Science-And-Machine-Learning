### **Selecting Data For Modeling**

Melbourn Housing Snapshot has too many variables, right now based on our intuition we will be picking up the variables and select to build this model

```python
melbourn_file_path = '/content/drive/MyDrive/Colab Notebooks/Data Science & Machine Learning/Data/melb_data.csv'
```

```python
import pandas as pd
```

```python
melbourn_data = pd.read_csv(melbourn_file_path)
```

To choose variables/columns, we'll neet to see a list of all columns in the dataset. That is done with the column property of the DataFrame

```python
melbourn_data.columns
```

The Melbourn Data has some missing values, So right now we will take the simplest way, Drop down the houses with missing values from our analysis.

dropna() function drops the missing values(think of na as 'not available')

```python
melbourn_data.dropna(axis=0)
```

There are many ways to select a subset of data, Right now we will focus on two approaches :
* Dot Notation, which we use to select the 'Prediction Target'
* Selecting with a column list, whicih we use to select the 'Features'

### Selecting The Prediction Target

A variable can be pulled out with dot-notation. This single column is stored in a Series, which is broadly like a DataFramewith only a single column of data.

Dor notation is used to select the column we want to predict, which is called prediction target, By convention, the prediction target is called as **y**.

```python
y = melbourn_data.Price
```

### Choosing 'Features'

The columns that are inputted into our model are called 'Features'. In our case, those would be the columns used to determine the home price. Sometimes, all columns are used except the target as features. other times you'll be better off with fewer features.

We select multiple features by providing a list of columns names inside brackets. Each item in that list should e a string.

By convention, this is called as **x**

```python
melbourn_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
x = melbourn_data[melbourn_features]
```

```python
# Reviewing the data we will use for our prediction

x.describe()
```

```python
# Viewing the few top rows

x.head()
```

#### **Building Model**

The steps to building and using a model are:

* **Define**: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
* **Fit**: Capture patterns from provided data. This is the heart of modeling.
* **Predict**: Just what it sounds like
* **Evaluate**: Determine how accurate the model's predictions are.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
melbourn_model = DecisionTreeRegressor(random_state=1)
```

```python
melbourn_model.fit(x, y)
```

Many machine learning models allow some randomness in model training. Specifying a number for random_state ensures you get the same results in each run. This is considered a good practice. You use any number, and model quality won't depend meaningfully on exactly what value you choose.

We now have a fitted model that we can use to make predictions.

In practice, you'll want to make predictions for new houses coming on the market rather than the houses we already have prices for. But we'll make predictions for the first few rows of the training data to see how the predict function works.

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(melbourn_model.predict(x.head()))
```
